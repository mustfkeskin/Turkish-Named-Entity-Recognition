```python
def writeDict(dict, filename, sep):
    with open(filename, "w", encoding="utf-8") as f:
        for key in dict.keys():            
            f.write("%s %s \n" % (key.replace(" ", "_"), ' '.join("%f" % val for val in dict[key])))
            
writeDict(meanVector, "../utils/meanVector.txt", " ")
writeDict(wordsDict, "../utils/wordsDict.txt", " ")
writeDict(all_dict, "../utils/allDict.txt", " ")
```

```python
import pandas as pd
df = pd.read_csv("../dbpedia/dbpedia20.csv")
```

```python
import gensim
filename = "../embeddings/cc.tr.w2v.bin"
word_vectors = gensim.models.KeyedVectors.load_word2vec_format(filename, binary=True)
w2v = dict(zip(word_vectors.wv.index2word, word_vectors.wv.syn0))
```

```python
th = 0.5
for i in range(df.shape[0]):
    words = df.loc[i, "vec"].replace('[', "").replace(']', "").replace("'", "").split()
    words = [word for word in words if word in w2v]
    doesnt_match = model_all.doesnt_match(words)
    
    new_label = [item for item in model_all.most_similar([doesnt_match] , topn=10)
                 if item[0] in model_mean.index2entity]
    
    if(len(new_label) > 0 ):
        #print(new_label[0])
        if(new_label[0][1] > th):
            new_label = new_label[0][0]
        else:
            new_label = ""
    else:
        new_label = ""
    #print(df.loc[i, "title"] + " , " + doesnt_match)
    print("title: " + df.loc[i, "title"] + " - " + doesnt_match + " : " + new_label)
```

```python
new_labels = {}
df.title = df.title.apply(lambda x : x.replace(" ", "_"))

for i in range(df.shape[0]):
    words = df.loc[i, "vec"].replace('[', "").replace(']', "").replace("'", "").split()
    words = [word for word in words if word in model_all]
    
    for word in words:
        new_label = [item for item in model_all.most_similar([word] , topn=10)
                     if item[0] in model_mean.index2entity]
        if(len(new_label) > 0 and new_label[0][1] > th):
            label = new_label[0][0]
            label = df.loc[df.title == label, "colors"].values[0]
            new_labels[word] = label + " --> " + new_label[0][0]
            print(word + " :  şu anki etiketi: " + label + " --> yeni etiketi: " + new_label[0][0])
```

```python
csv = pd.DataFrame()
csv["word"] = new_labels.keys()
csv["label"] = new_labels.values()

csv["colors"] = csv["label"].apply(lambda x : x.split(" --> ")[0])
csv["title"] = csv["label"].apply(lambda x : x.split(" --> ")[1])
csv = csv.drop(["label"], axis = 1)
csv = csv.sort_values(["colors","title"])
csv = csv.reset_index(drop=True)

csv.to_csv("../utils/csv.csv", index=False)
```
